# **Historia**

El desarrollo histórico de las Interfaces Cerebro-Computador (BCI) representa la transición de especulaciones teóricas hacia una rigurosa disciplina clínica y neurotecnológica. Los cimientos de esta tecnología se establecieron en 1924, cuando el neurólogo alemán Hans Berger registró por primera vez la actividad eléctrica del cerebro humano mediante electroencefalografía (EEG), demostrando que las fluctuaciones neuronales podían capturarse de forma no invasiva para su posterior estudio. Sin embargo, no fue hasta 1973 que el investigador Jacques Vidal acuñó formalmente el término "Interfaz Cerebro-Computador", definiendo un sistema experimental capaz de utilizar las señales neuronales en un diálogo directo persona-computadora, el cual funciona como una extensión protésica del cerebro sin requerir el uso de las vías de salida neuromusculares o periféricas tradicionales [1]. Durante las décadas de 1990 y 2000, los avances sustanciales en la capacidad de procesamiento informático y en los algoritmos matemáticos de clasificación de señales permitieron a pioneros como Jonathan Wolpaw desarrollar aplicaciones prácticas, tales como sistemas de deletreo basados en EEG y las primeras matrices microelectródicas intracorticales que demostraron ser capaces de registrar potenciales de acción de neuronas individuales para el control primario de cursores y dispositivos mecánicos externos [1].

# **Uso**

En su uso clínico actual, la tecnología BCI se posiciona como un pilar fundamental en las unidades de neurorrehabilitación y asistencia tecnológica, brindando soluciones tangibles para pacientes con discapacidades motoras profundas derivadas de condiciones como la Esclerosis Lateral Amiotrófica (ELA), lesiones medulares severas o accidentes cerebrovasculares (ACV). A nivel operativo, los sistemas BCI decodifican oscilaciones cerebrales específicas, como los ritmos sensoriomotores, y las traducen en comandos de salida que permiten a los pacientes operar sillas de ruedas eléctricas, exoesqueletos ortopédicos o interfaces de software para comunicación aumentativa. En el ámbito específico de la rehabilitación motora posictus, metaanálisis recientes han comprobado que el uso repetitivo de BCI proporciona una retroalimentación sensorial (visual o propioceptiva) inmediata de las oscilaciones cerebrales en curso, permitiendo a los sobrevivientes modular a voluntad sus propios ritmos motores [2]. Esta interacción sostenida entre la intención mental y la respuesta física del dispositivo no solo induce mejoras clínicas significativas, cuantificables en escalas estandarizadas como la Evaluación de Fugl-Meyer para miembros superiores (FMA-UE), sino que desencadena potentes mecanismos de neuroplasticidad estructural y funcional a nivel subclínico, reconectando las vías neuronales residuales que fueron dañadas por la falta de oxígeno [2].



# **Proyectos futuros**

El desarrollo de proyectos futuros en el ámbito de las Interfaces Cerebro-Computador (BCI) se orienta hacia la restitución de capacidades humanas altamente complejas y a la drástica reducción del riesgo quirúrgico asociado a los implantes cerebrales. En lugar de limitarse al control básico de cursores o extremidades robóticas lentas, las investigaciones emergentes se centran en decodificar procesos cognitivos rápidos (como la fonética del habla), utilizar vías de acceso anatómicas alternativas para no perforar el cráneo, y diseñar materiales biocompatibles que se adapten a la anatomía cerebral sin generar rechazo. Estas iniciativas marcan la transición de pruebas de concepto experimentales hacia dispositivos médicos viables a largo plazo. A continuación, se describen tres proyectos disruptivos con sus respectivos enfoques y bases tecnológicas:

**1) Neuroprótesis para la Decodificación del Habla**

Este proyecto se basa en devolver la capacidad de comunicación a velocidad conversacional a pacientes con síndrome de enclaustramiento o parálisis severa. A diferencia de las BCI tradicionales que requieren que el paciente seleccione letras una por una en una pantalla (deletreo), esta tecnología se fundamenta en implantar matrices de electrodos directamente en la corteza motora del habla. El sistema decodifica la actividad cerebral asociada a los intentos de realizar movimientos articulatorios microscópicos (labios, lengua, mandíbula y laringe) y utiliza algoritmos de aprendizaje automático para traducir instantáneamente esas señales en texto o en una voz sintetizada, logrando velocidades que superan las 60 palabras por minuto [5].

**2) Interfaces Endovasculares (Proyecto Stentrode)**

Esta iniciativa busca resolver el problema de la invasividad y el trauma de las cirugías a cerebro abierto (craneotomías). El proyecto se basa en la neurotecnología endovascular, la cual utiliza la red de vasos sanguíneos del propio cerebro como vía de acceso. Se introduce un arreglo de electrodos montado sobre un stent (una malla metálica expansible) a través de la vena yugular en el cuello, navegándolo hasta llegar al seno sagital superior, justo por encima de la corteza motora. Una vez allí, el dispositivo se expande, se fusiona con las paredes del vaso sanguíneo y comienza a registrar de manera segura las señales eléctricas del cerebro para controlar dispositivos digitales sin que los electrodos toquen directamente el tejido neuronal [6].


 **Electrodos Flexibles y Nanomateriales de Interfaz (Electrónica Blanda)**

Este proyecto se basa en resolver el problema crítico del rechazo inmunológico y la degradación de la señal a largo plazo que sufren los implantes cerebrales tradicionales. El cerebro humano es un tejido extremadamente suave que pulsa y se mueve de manera natural. Cuando se insertan electrodos rígidos (como los de silicio o metal), la constante fricción causa una respuesta inflamatoria y la formación de una cicatriz alrededor del implante (gliosis) que termina bloqueando la señal eléctrica. La solución actual se fundamenta en el desarrollo de "electrodos blandos" utilizando polímeros orgánicos ultrafinos, elásticos y biocompatibles cuyas propiedades mecánicas imitan la suavidad del propio tejido cerebral. Además, la superficie de estos electrodos se modifica con nanomateriales conductores (como el polímero PEDOT:PSS o biopolímeros). Esto permite que el dispositivo se mueva al mismo ritmo que las pulsaciones del cerebro sin causar daño mecánico, reduciendo la respuesta inmunitaria casi a cero y garantizando que la Interfaz Cerebro-Computadora pueda registrar señales de alta fidelidad durante años [7].

**Prótesis Visuales Corticales para la Restauración de la Visión**

Este proyecto se basa en devolver un grado funcional de percepción visual a personas con ceguera total adquirida, eludiendo por completo los ojos, la retina y el nervio óptico dañados. Su funcionamiento se fundamenta en la implantación de matrices de microelectrodos directamente en la corteza visual (ubicada en el lóbulo occipital del cerebro). El sistema utiliza unas gafas equipadas con una cámara externa que captura el entorno en tiempo real; luego, un microprocesador traduce esas imágenes en patrones específicos de estimulación eléctrica. Al inyectar estas señales directamente en el tejido cortical, se induce en el cerebro la percepción de "fosfenos" (destellos de luz artificiales), lo que permite a los pacientes identificar contornos, reconocer formas básicas y evadir obstáculos para navegar por su entorno de forma más independiente [8].

# **3) Actividad cerebral registrada por EEG durante movimientos e imaginación de movimiento de las manos**


Las Interfaces Cerebro-Computador (BCI) han dejado de ser ciencia ficción para convertirse en una realidad clínica e ingenieril capaz de traducir la intención mental pura en comandos digitales. Uno de los paradigmas más estudiados en este campo es la Imaginación Motora (MI), un proceso neurológico donde el simple acto de pensar en mover las manos genera alteraciones rítmicas medibles en la actividad eléctrica de la corteza cerebral. Mediante el registro no invasivo a través de electroencefalografía (EEG) y la aplicación de sofisticados algoritmos de inteligencia artificial, los sistemas modernos logran decodificar estos sutiles patrones en tiempo real. A continuación, se presenta un análisis detallado sobre las bases neurofisiológicas, las técnicas de procesamiento de señales y la comparación de los modelos de clasificación.

# **1. Ritmos Cerebrales de Interés (SMR - Sensorimotor Rhythms)**

Tanto la ejecución de un movimiento real de las manos como la imaginación del mismo (pensar en moverlas sin contraer los músculos) activan la corteza sensoriomotora del cerebro. Los sistemas BCI analizan fenómenos oscilatorios específicos conocidos como Desincronización Relacionada con Eventos (ERD) y Sincronización Relacionada con Eventos (ERS) en dos bandas de frecuencia principales [9], [10]:

**1) Ritmo Mu (μ):** Oscila en la banda de 8 a 12 Hz (comparte frecuencia con el ritmo Alfa, pero se localiza exclusivamente sobre la corteza motora). Cuando una persona está en reposo, este ritmo está sincronizado (alta amplitud). Sin embargo, cuando la persona mueve la mano o imagina que la mueve, ocurre una caída drástica en la amplitud de la señal. A esto se le llama ERD (Desincronización Relacionada con el Evento) [9], [10].

**2) Ritmo Beta (β):** Oscila entre los 13 y 30 Hz. Al igual que el ritmo Mu, sufre una atenuación (ERD) durante la preparación y ejecución/imaginación del movimiento. Sin embargo, su característica más importante es el "Rebote Beta" (Beta Rebound): una vez que el movimiento o la imaginación cesa, hay un aumento transitorio y pronunciado en la amplitud de la señal (ERS), indicando que la red neuronal motora se está "reiniciando" o inhibiendo [11].


# **2. Adquisición de Señales: El Rol de los Electrodos EEG**

Para capturar estos cambios sutiles de voltaje (que se miden en microvoltios, $\mu V$), el hardware debe superar la resistencia natural de la piel y el cráneo.

**Posicionamiento Espacial (Contralateralidad):**
El cerebro opera de forma cruzada. Al imaginar el movimiento de la mano derecha, la ERD se manifiesta de manera más pronunciada en el hemisferio izquierdo de la corteza motora. Por ello, según el Sistema Internacional 10-20, los electrodos críticos para decodificar las manos son el C3 (hemisferio izquierdo, mano derecha) y el C4 (hemisferio derecho, mano izquierda), frecuentemente apoyados por el electrodo Cz como referencia [12].

**Tecnología de Electrodos:**

 **Electrodos Húmedos (Ag/AgCl)**: Utilizan un gel electrolítico abrasivo que reduce la impedancia de la piel (típicamente < 5 k$\Omega$), permitiendo la captura de señales de alta fidelidad (el estándar clínico). Sin embargo, el gel se seca con el tiempo, degradando la señal.

**Electrodos Secos:** Utilizan pines de contacto directo. Aunque permiten una configuración rápida (ideal para BCI de uso diario), sufren de una impedancia muy alta y son sumamente susceptibles a capturar ruido eléctrico ambiental y artefactos musculares, exigiendo un preprocesamiento algorítmico mucho más severo [13].


# **3)Tubería de procesamientos de señales BCI**

**1. Adquisición de Señales (Hardware)**

Esta fase corresponde a la captura física de la actividad cerebral superficial mediante electrodos (típicamente siguiendo el sistema internacional 10-20). Dependiendo del objetivo clínico o de ingeniería, los investigadores seleccionan el tipo de sensor (húmedo de Ag/AgCl o seco), la frecuencia de muestreo (cuántas mediciones por segundo toma el amplificador, generalmente entre 100 Hz y 1000 Hz) y la densidad espacial (desde 2 únicos electrodos críticos como C3 y C4, hasta mallas de alta densidad de 118 o 128 canales).

**2. Preprocesamiento y Limpieza (Acondicionamiento de Señal)**

La señal cruda de EEG no puede usarse directamente porque está contaminada. El preprocesamiento tiene dos misiones: delimitar la información cerebral útil y aislarla del "ruido" físico.  Filtrado de Frecuencia: Se aplican filtros matemáticos (como pasabanda o filtros Butterworth) para descartar toda la actividad eléctrica que no pertenece a los ritmos sensoriomotores de interés. Generalmente, se aíslan las frecuencias correspondientes a las bandas Mu (8-12 Hz) y Beta (13-30 Hz).  Eliminación de Artefactos: Los electrodos captan inadvertidamente el voltaje del corazón (ECG), los parpadeos (EOG) y la tensión muscular de la mandíbula o el cuello (EMG). Para limpiar el EEG, se emplean algoritmos que rechazan los canales ruidosos o métodos estadísticos avanzados como el Análisis de Componentes Independientes (ICA), el cual separa la señal cerebral de los parpadeos sin perder los datos de la imaginación motora.

**3. Extracción de Características (Feature Extraction)**

Una vez que la señal está limpia y enfocada en las bandas frecuenciales correctas, el sistema debe encontrar un "patrón" numérico que defina qué es "mano izquierda" y qué es "mano derecha". Esta etapa transforma las ondas oscilatorias en vectores matemáticos.En los enfoques tradicionales, los ingenieros utilizan algoritmos en el dominio espacial (creando filtros matemáticos que maximizan las diferencias de amplitud entre hemisferios) o en el dominio tiempo-frecuencia (evaluando la energía de la señal en milisegundos específicos).En los enfoques de aprendizaje profundo (Deep Learning), esta fase manual desaparece; la red neuronal recibe la señal preprocesada y descubre matemáticamente las características por sí sola.  

**4. Clasificación (Machine Learning / Decodificación)**

En la última etapa de la tubería, un algoritmo de inteligencia artificial recibe el vector de características extraído en el paso anterior y traza una frontera de decisión geométrica o probabilística para asignar una etiqueta final. El clasificador evalúa los datos y emite el comando de salida ("Clase 1: Mano Izquierda", "Clase 2: Mano Derecha"). El rendimiento de esta etapa define la precisión total de la Interfaz Cerebro-Computador.

 Al revisar la literatura para estructurar este proyecto de Interfaz Cerebro-Computador (BCI), se observa que los autores proponen principalmente dos grandes enfoques para procesar las señales de imaginación motora, además de un par de soluciones para llevar el sistema a un entorno práctico. Primero está el procesamiento tradicional basado en Machine Learning clásico. En este método, el trabajo pesado se realiza manualmente en la etapa de extracción de características, usando algoritmos matemáticos como los Filtros Espaciales Comunes (CSP) para forzar y resaltar las diferencias de voltaje entre el hemisferio derecho y el izquierdo del cerebro. La ventaja de este enfoque es que resulta muy fácil de interpretar y no requiere una computadora muy potente. Sin embargo, al analizar los estudios, se evidencia que es un método muy sensible al ruido; si el usuario aprieta la mandíbula o parpadea, los clasificadores tradicionales suelen confundirse y fallar.

Por otro lado, la tendencia que domina los artículos más recientes es el uso de Deep Learning. En lugar de extraer las características a mano mediante ensayo y error, los autores proponen usar Redes Neuronales Convolucionales (CNN) o modelos recurrentes que reciben la señal preprocesada directamente y aprenden los patrones por sí solos. Esta opción resulta mucho más atractiva porque las redes logran precisiones más altas y encuentran relaciones complejas en el EEG que los métodos clásicos pasan por alto. El gran reto con esta alternativa es que funciona como una "caja negra", por lo que es difícil explicar exactamente por qué la red tomó cierta decisión, y además requiere muchísimos datos de entrenamiento para que el modelo funcione bien y no se sobreajuste.

Justamente por la necesidad de tantos datos, otro tipo de procesamiento que se discute bastante para este tipo de proyectos es la Transferencia de Aprendizaje (Transfer Learning). Uno de los mayores problemas prácticos que tienen los sistemas BCI es que las ondas cerebrales varían muchísimo de una persona a otra, así que un modelo entrenado con un usuario no le sirve a otro. Para solucionar esto, las propuestas sugieren tomar un modelo que ya fue preentrenado con bases de datos públicas enormes y simplemente hacerle un ajuste fino o calibración rápida con el nuevo usuario. Esto es fundamental porque evitaría que la persona tenga que pasar horas imaginando movimientos para entrenar el sistema desde cero cada vez que se lo pone.

Finalmente, pensando en la implementación física del proyecto, destacan las propuestas enfocadas en el procesamiento embebido (Edge Computing). Cuando se busca pasar de una simulación en la computadora a controlar algo en la vida real, como un exoesqueleto o un cursor, la prioridad cambia. Ya no solo importa tener una precisión perfecta, sino que el procesamiento debe ocurrir en milisegundos para evitar retrasos (latencia). Por eso, varios autores plantean optimizar y comprimir las redes neuronales para que puedan correr directamente en microprocesadores pequeños montados en el hardware del usuario, garantizando que el comando mental se sienta natural. En conclusión, para que un proyecto de este tipo sea viable y moderno, la mejor ruta parece ser la integración de modelos de Deep Learning por su precisión, apoyándose en técnicas de Transfer Learning para resolver el problema de la variabilidad entre las personas.

###  METODOLOGÍA DEL PROYECTO

**1. Adquisición de la Señal**

Para el desarrollo de este proyecto, se utilizó la base de datos de acceso público *EEG Motor Movement/Imagery Dataset* proporcionada por PhysioNet, la cual fue registrada mediante el sistema BCI2000 [24]. El análisis se llevó a cabo utilizando la totalidad del conjunto de datos, abarcando los registros de 109 sujetos experimentales (`sub-001` a `sub-109`). De los diferentes bloques experimentales disponibles, se seleccionaron y concatenaron específicamente las tareas correspondientes a la ejecución del movimiento real (corridas 3, 7 y 11) y a la imaginación motora (corridas 4, 8 y 12) de ambas manos. Dado que la actividad neuronal vinculada al movimiento se localiza principalmente en la corteza motora primaria, se realizó una selección espacial aislando únicamente las señales registradas por los electrodos C3 (hemisferio izquierdo) y C4 (hemisferio derecho) del sistema internacional 10-20.

**2. Preprocesamiento**

La señal electroencefalográfica cruda contiene diversas fuentes de ruido que deben ser atenuadas. Para acondicionar los datos, se implementó mediante la librería MNE-Python un filtro digital pasa-banda tipo FIR (*Finite Impulse Response*) utilizando una ventana Hamming, con frecuencias de corte establecidas entre 8 Hz y 30 Hz. La selección de este rango específico de frecuencias se fundamenta teóricamente en la necesidad de aislar los ritmos sensoriomotores, específicamente las bandas mu (8-12 Hz) y beta (13-30 Hz). Estas bandas son las que experimentan el fenómeno fisiológico conocido como Desincronización Relacionada con Eventos (ERD) durante la planificación y ejecución (real o imaginaria) de tareas motoras [9],[25].

**3. Extracción de Características y Análisis Estadístico**

Tras el filtrado de las señales espaciales y temporales, se procedió a la extracción de características mediante el cálculo de la Potencia Relativa de Banda (PRB) para cuantificar la energía presente en las bandas mu y beta de los canales C3 y C4. Para validar la hipótesis de que existe una activación contralateral medible durante las tareas de imaginación y movimiento (por ejemplo, mayor actividad en C4 al mover la mano izquierda), se diseñó un flujo de análisis estadístico utilizando el paquete `scipy.stats`. Inicialmente, se evaluó la normalidad de las distribuciones de los datos extraídos aplicando la prueba de Shapiro-Wilk [26]. Dependiendo del cumplimiento del supuesto de normalidad, las diferencias de potencia entre los electrodos fueron sometidas a pruebas de hipótesis emparejadas: se aplicó la prueba paramétrica T de Student para muestras relacionadas ($T-test$) cuando los datos exhibían una distribución normal, y la prueba no paramétrica de rangos con signo de Wilcoxon cuando los datos no presentaban normalidad.
